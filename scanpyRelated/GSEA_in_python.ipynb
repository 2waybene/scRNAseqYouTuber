```python
import pandas as pd
from gseapy.plot import gseaplot
import gseapy as gp
import numpy as np
```

```python
df = pd.read_csv('DE_results.csv').dropna()
df = df.rename(columns = {'symbol': 'Gene'})
#df = df[df.padj < 0.05] it is best practice not to filter

df
```

```python
df['Rank'] = -np.log10(df.padj)*df.log2FoldChange

df = df.sort_values('Rank', ascending = False).reset_index(drop = True)

df
```

```python
ranking = df[['Gene', 'Rank']]
ranking
```

```python
gp.get_library_name()
```

```python
pre_res = gp.prerank(rnk = ranking, gene_sets = 'GO_Biological_Process_2021', seed = 6, permutation_num = 100)
```

```python
out = []

for term in list(pre_res.results):
    out.append([term,
               pre_res.results[term]['fdr'],
               pre_res.results[term]['es'],
               pre_res.results[term]['nes']])

out_df = pd.DataFrame(out, columns = ['Term','fdr', 'es', 'nes']).sort_values('fdr').reset_index(drop = True)
out_df
```

```python
term_to_graph = out_df.iloc[0].Term
term_to_graph
```

```python
gseaplot(pre_res.ranking, term = term_to_graph, **pre_res.results[term_to_graph])
```

```python
#this block is just me making a enriched set of genes in the DE upregulated genes
#not required for GSEA
import random

def a_thing(x):
    if x < 0:
        return 1
    else:
        return x
df['weights'] = df.Rank.map(a_thing)

a_list = random.choices(df.Gene.values, k = 100, weights = df.weights.values)
```

```python
gseaplot(pre_res.ranking, term = 'set_name1', **pre_res.results['set_name1']) #ofname = 'saved_fig.png'
```

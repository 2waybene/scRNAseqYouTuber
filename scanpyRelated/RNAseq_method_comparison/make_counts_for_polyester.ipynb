```python
#picking 5 samples with similar number of total reads for the replicates. 5 closest samples to the q75 value

sums = df.sum()

sums = pd.to_numeric(sums, errors='coerce')
sums = sums.dropna()
sums = sums.sort_values()

#median_value = sums.median() #changed it to q75 instead of median
median_value = sums.quantile(0.75)

differences = abs(sums - median_value)

sorted_diffs = differences.sort_values()

closest_values = sums[sorted_diffs.index].head(5)
print(closest_values)
```

```python
closest_values.index
```

```python
#create the counts

columns = []

for x in [1,5,10,15,20,25,30,40,50,60,70,80,90,100]:
    p = x/100

    for samp in closest_values.index:
    
        a = df[samp].copy()
    
        a = ((a*p))
        a[a < 1] = 0

        a = a.round()
        
        columns.append(a)
```

```python
#output names to match the polyster default output na
thinger = pd.DataFrame(np.column_stack(columns), columns=[f"sample_{j:02}" for j in range(1, len(columns) + 1)])
```
